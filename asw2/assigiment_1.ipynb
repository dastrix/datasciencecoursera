```python
from pprint import pprint

f = open('sentences.txt')
fdata = f.readlines()
lines = [line.lower() for line in fdata]
wlines = [line.split() for line in lines]

words = []
[words.extend(line) for line in wlines]
words = list(set(words))
words = filter(None, words)
# print words

dct = {y:x for x,y in enumerate(words)}

matrix = [[0 for i in range(len(words))] for j in range(len(lines))]

i = 0
for word in words:
    j = 0
    for wline in wlines:
        matrix[j][i] = wline.count(word)
        j+=1
    i+=1
    
print len(matrix), len(matrix[0])
```

```python
# import scipy
from scipy.spatial import distance
import operator

answ = {}
i = 0
for row in matrix:
    answ[i] = scipy.spatial.distance.cosine(matrix[0], row)
    i+=1;
print answ
print sorted(answ.items(), key=operator.itemgetter(1))
```
